####1. 
Approximate $f(x)=x^{0.321}$ with a Taylor series around $\bar x = 1$. Compare your approximation over the domain (0,4). Compare when you use up to 1, 2, 5 and 20 order approximations. Discuss your results

```python
import sympy as sy
import numpy as np
import matplotlib.pyplot as plt

plt.style.use("ggplot")

# Define the variable and the function to approximate
x = sy.Symbol('x')
def f(x):
	return x**0.321


# Factorial function
def factorial(n):
    if n <= 0:
        return 1
    else:
        return n*factorial(n-1)

# Taylor approximation at x0 of the function 'function'
def taylor(function,x0,n):
    i = 0
    p = 0
    while i <= n:
        p = p + (function.diff(x,i).subs(x,x0))/(factorial(i))*(x-x0)**i
        i += 1
    return p
```

```python
# Compare when you use up to 1, 2, 5 and 20 order approximations. Discuss your results.

# Plot results
def plot():
    x_lims = [0,4]
    x1 = np.linspace(x_lims[0],x_lims[1],800)
    y1 = []
    # Approximate up until 10 starting from 1 and using steps of 2
    m1 = [1, 2, 5, 20]
    for j in m1:
        func = taylor(f(x),1,j)
        print('Taylor expansion at n='+str(j),func)
        for k in x1:
            y1.append(func.subs(x,k))
        plt.plot(x1,y1,label='order '+str(j))
        y1 = []
    # Plot the function to approximate
    plt.plot(x1,f(x1),label='$x^{0.321}$')
    plt.xlim(x_lims)
    plt.ylim([0,4])
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    plt.grid(True)
    plt.title('Taylor series approximation')
    plt.show()

plot()
```

Above Results show that for function which has a singularity, in our case, at $x =  0$. Taylor approximate $f$ at $\bar x = 1$, 
the approximation error of this approximation increases sharply for $x > 1$ and the error is getting bigger (deviate from the original function) when approximate order is bigger as shown in our Figure.

```python
# Report the error graph:
def plot():
    x_lims = [0,4]
    x1 = np.linspace(x_lims[0],x_lims[1],800)
    y1 = []
    # Approximate up until 10 starting from 1 and using steps of 2
    m1 = [1, 2, 5, 20]
    for j in m1:
        func = taylor(f(x),1,j)
        print('Taylor expansion at n='+str(j),func)
        for k in x1:
            y1.append(func.subs(x,k))
        plt.plot(x1,y1-f(x1),label='order '+str(j))
        y1 = []
    # Plot the function to approximate
    plt.xlim(x_lims)
    plt.ylim([0,4])
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    plt.grid(True)
    plt.title('Taylor series approximation')
    plt.show()

plot()
```

#### 3.
Approximate these function: $e^{1/x}$, the runge function $\frac{1}{1+25x^2}$, and the ramp function $f(x)=\frac{x+|x|}{2}$ for the domain $x \in [-1,1]$ with:

$$ x_i = a + \frac{i-1}{n-1}(b-a) $$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import CubicSpline
import numpy.polynomial.polynomial as poly
from numpy import inf
# Runge's function in [-1,1]
def f(x):
	return 1.0 / (1.0 + 25.0 * x**2)

# n evenly spaced interpolation nodes:
x_lims=[-1,1]
x1 = np.linspace(x_lims[0],x_lims[1],11) 
y = f(x1)
# Or:
# x = np.arange(-1,1,.1)

xs = np.linspace(x_lims[0],x_lims[1],101)
cs = poly.polyfit(x1,y,3)
ffit3 = poly.polyval(xs,cs)
# Monomials of order 5:
coeffs = poly.polyfit(x1,y,5)
ffit = poly.polyval(xs,coeffs)

# Monomials of order 10: 
coeffs10 = poly.polyfit(x1,y,10)
ffit10 = poly.polyval(xs,coeffs10)

# Exponential Function
def g(x):
	with np.errstate(divide='ignore', invalid='ignore'):
		return np.exp(1/x)

yg = g(x1)
yg[yg == inf] = 9999


csg = poly.polyfit(x1,yg,3)
coeffg = poly.polyfit(x1,yg,5)

ffitg = poly.polyval(xs,coeffg)
ffit3g = poly.polyval(xs,csg)

# Monomials of order 10: 
coeffs10g = poly.polyfit(x1,yg,10)
ffit10g = poly.polyval(xs,coeffs10g)


def h(x):
	return .5*(x+abs(x))
yh = h(x1)

csh = poly.polyfit(x1,yh,3)
ffit3h = poly.polyval(xs,csh)
# Monomials of order 5:
coeffsh = poly.polyfit(x1,yh,5)
ffith = poly.polyval(xs,coeffsh)

# Monomials of order 10: 
coeffs10h = poly.polyfit(x1,yh,10)
ffit10h = poly.polyval(xs,coeffs10h)


# plot
# plot with various axes scales
fig = plt.figure(figsize=(10,10))
fig.suptitle('Function Approximation: Evenly-Spaced nodes & monomials')

ax1 = plt.subplot(321)
plt.plot(x1,y,'o',label='data')
plt.plot(xs,f(xs), label='True')
plt.plot(xs,ffit3,label='Cubic Polynomials') 
plt.plot(xs,ffit,label='Monomial of order 5')
plt.plot(xs,ffit10,label='Monomial of order 10')
plt.legend(bbox_to_anchor=(0.5,0), loc="lower right", bbox_transform=fig.transFigure, ncol=3)
plt.ylim(-0.2,1)
plt.title('Runge Function $1 / (1 + 25x^2) $')

plt.subplot(322, sharex=ax1)
plt.plot(xs,f(xs)-ffit3,label='Cubic error') 
plt.plot(xs,f(xs)-ffit,label='m5 error')
plt.plot(xs,f(xs)-ffit10,label='m10 error')
plt.ylim(0,1)
plt.legend(bbox_to_anchor=(1,0), loc="lower right", bbox_transform=fig.transFigure, ncol=3)
plt.title('Runge Function Approximation Errors')

plt.subplot(323, sharex=ax1)
plt.plot(x1,yg,'o')
plt.plot(xs,g(xs))
plt.plot(xs,ffit3g) 
plt.plot(xs,ffitg)
plt.plot(xs,ffit10g)
plt.ylim(-100,1000)
plt.title('Exponential Function $exp(1/x)$')

plt.subplot(324, sharex=ax1)
plt.plot(xs,g(xs)-ffit3h) 
plt.plot(xs,g(xs)-ffitg)
plt.plot(xs,g(xs)-ffit10g)
plt.ylim(0,100)
plt.title('Exponential Function Approximation Errors')


plt.subplot(325, sharex=ax1)
plt.plot(x1,yh,'o')
plt.plot(xs,h(xs))
plt.plot(xs,ffit3h) 
plt.plot(xs,ffith)
plt.plot(xs,ffit10h)
plt.ylim(-0.5,2.5)
plt.title('Ramp Function $(x+|x|)/2$')

plt.subplot(326, sharex=ax1)
plt.plot(xs,h(xs)-ffit3h) 
plt.plot(xs,h(xs)-ffith)
plt.plot(xs,h(xs)-ffit10h)
plt.ylim(0,0.25)
plt.title('Ramp Function Approximation Errors')

plt.xlim(x_lims)
# Adjust the subplot layout, because the logit one may take more space
# than usual, due to y-tick labels like "1 - 10^{-3}"
plt.subplots_adjust(top=0.92, bottom=0.08, left=0.10, right=0.95, hspace=0.25,wspace=0.35)
plt.show()
```

$$ \sigma_{1,2} = \frac{dln(h/k)}{dln(MRS_{1,2})}$$

where 

$$MRS_{1,2} = \frac{\partial f(k,h)/ \partial k}{\partial f(k,h) / \partial h}$$ 
$$ = \frac{1-\alpha}{\alpha}(k/h)^{\frac{-1}{\sigma}}$$

compute elasticity of substitution of CES: 

$$\sigma_{1,2} = \frac{d \ln (h/k)}{d \ln (MRS_{1,2})}$$
$$ = \frac{d \ln (h/k)}{d \ln (\frac{1-\alpha}{\alpha}(k/h)^{\frac{-1}{\sigma}}) }$$

Above equation, regard $\ln (h/k) = \ln m $, then, it changes to:

$$\sigma_{1,2} = \frac{d \ln m}{d [\ln (1-\alpha)/(\alpha) + (-1/\sigma \ln m)]}$$

$$\sigma_{1,2} = - \sigma$$

Since, sometimes, we prefer the define elasticity of substitution with a minus above, thus, the Elatisicity of substitution in the CES is just $\sigma$



- Compute labor share for an economy with that CES production function assuming factor inputs face competitive markets.

The definition of labor share is: 
$$ \frac{h w}{f(k,h)}$$ 
, where, $w$ is the wage obtained from Competitive markets, in this case is:

$$w = \alpha h^{-1/\sigma} ((1-\alpha) k^{\frac{\sigma-1}{\sigma}} + \alpha h^{\frac{\sigma-1}{\sigma}})^{1/(\sigma-1)} $$

Then, 

$$ \frac{h w}{f(k,h)} = \frac{\alpha h^{(\sigma - 1 )/ \sigma }}{((1-\alpha) k^{\frac{\sigma-1}{\sigma}} + \alpha h^{\frac{\sigma-1}{\sigma}})^{\sigma}}$$ 

### Chebyshev Approximation Algorithm in $R^2$
- Objective: Given $f(x,y)$ defined on [a,b]  [c,d], find its Chebyshev polynomial approximation $p(x, y)$

- Step 1: Compute the $m \geq n + 1$ Chebyshev interpolation nodes on [1, 1]:
$$z_k = -cos(\frac{2k-1}{2m} \pi), k=1, ..., m$$

- Step 2: Adjust nodes to [a, b] and [c, d] intervals:
$$x_k = (z_k + 1)(\frac{b-a}{2})+a, k=1, ...,m$$
$$y_k = (z_k + 1)(\frac{d-c}{2})+c, k=1, ...,m$$

- Step 3: Evaluate $f$ at approximation nodes:
$$w_{k,l}=f(x_k,y_l), k = 1, ... , m, l= 1, ..., m $$

- Step 4: Compute Chebyshev coefficients, $a_{ij}$, $i, j = 0,    , n$:

$$a_{ij}=\frac{\sum_{k=1}^{m} \sum_{l=1}^{m}w_{k,l} T_{i} (z_k) T_{j} (z_l)}{(\sum_{k=1}^{m} T_{i} (z_k)^2)(\sum_{l=1}^{m} T_{j} (z_l)^2)} $$

to arrive at approximation of $f(x,y)$ on [a,b]  [c,d]:
$$ p(x,y)= \sum_{i=0}^{n} \sum_{j=0}^{n} a_{ij} T_i (2\frac{x-a}{b-a}-1)T_j (2\frac{y-c}{d-c}-1) $$

```python
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm

# CES
def f(alpha,rho,k,h):
	return np.power((1-alpha)*k**rho + alpha*h**rho,1/rho)

alpha = .5
sigma = .25
rho= (sigma -1)/sigma

# Generate the nodes by above procedures: 
N = 20
m1 = np.cos((2*np.arange(1,N+1)-1)/(2*N)*np.pi) 
# shifting to domain (0,10] since they are the same
q1 = 5*(1+m1)
k1 = 5*(1+m1)
# k1 = sorted(k1)
k1 = np.asarray(k1)
h1 = 5*(1+m1)
h1 = np.asarray(h1)
q1 = np.asarray(q1)
m1 = np.asarray(m1)
# k_lims=[0,10]
# h_lims=[0,10]
# ks = np.linspace(k_lims[0],k_lims[1],101)
# hs = np.linspace(h_lims[0],h_lims[1],101)

f_nodes = f(alpha,rho,k1[:,None],h1[None,:])
f_nodes = np.matrix(f_nodes)
```
